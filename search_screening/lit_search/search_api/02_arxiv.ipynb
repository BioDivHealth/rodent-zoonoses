```python
import numpy as np
import pandas as pd
import arxiv
```

```python
import arxiv

client = arxiv.Client()

search = arxiv.Search(
  query = "all:'rodent trap' AND all:'africa'",
  max_results = 1000,
  sort_by = arxiv.SortCriterion.Relevance
)

results = client.results(search)
```

```python
link = []
authors = []
title = []
keywords = []
publication_date = []
abstract = []
journal = []
categories = []

for article in results:
    link.append(article.entry_id)
    authors.append(article.authors)
    title.append(article.title)
    #if article.keywords:
       # if None in article.keywords:
         #   article.keywords.remove(None)
        #keywords = '", "'.join(article.keywords)
    publication_date.append(article.published)
    abstract.append(article.summary)
   # categories.append(article.categories)
    #journal.append(article.journal)


df = pd.DataFrame({'link':link,
                   'authors':authors,
                   'title':title,
                   #'keywords':keywords,
                   'publication_date':publication_date,
                   #'categories':categories,
                   'abstract':abstract#,'journal':journal
                  })
```
